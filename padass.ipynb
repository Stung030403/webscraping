Merge, Join, Concattenate

```python
df1= pd.read_csv('/Users/mac/tung/python/LOTR.csv')
df1
```

```python
df2= pd.read_csv('/Users/mac/tung/python/LOTR 2.csv')
df2
```

```python
df1.merge(df2) #same data in 2 df
df1.merge
```

```python
df1.merge(df2,how='outer') #merge all data, inner is automatic

```

```python
df1.merge(df2,how= 'left')# keep all rows in df1(left data frame) and matched data in df2
```

```python
df2.merge(df2,how='right')# keep all rows in df2(right data frame) and matched data in df1
```

```python
df1.merge(df2,how= 'cross') #compare each row in df1 to every row in df2
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```

merge, join, concat
1. merge()
Purpose: Combines two DataFrames based on one or more keys (columns)
2. join()
Purpose: Combines two DataFrames based on their indexes or a key column.
3. concat()
Purpose: Concatenates two or more DataFrames along a particular axis (rows or columns).
stacks DataFrames either vertically (along rows) or horizontally (along columns).
